### Reaction Time between dispatch of unit and reaching on scene

```python
# Convert the datetime columns to datetime objects
df['Dispatch DtTm'] = pd.to_datetime(df['Dispatch DtTm'])
df['On Scene DtTm'] = pd.to_datetime(df['On Scene DtTm'])

# Calculate the time difference in minutes between the two columns
df['reaction_time'] = (df['On Scene DtTm'] - df['Dispatch DtTm']).dt.total_seconds()
df['reaction_time'] = df['reaction_time'] / 60

# Print the first 3 rows of the DataFrame with the new column
print(df.head(3))
```

```python
reaction_time_min = df['reaction_time'].min()
reaction_time_max = df['reaction_time'].max()
```

```python
def classify_reaction_time(row):
    if row['reaction_time'] <= reaction_time_min + (reaction_time_max - reaction_time_min) / 3:
        return 'Early'
    elif row['reaction_time'] <= reaction_time_min + 3 * (reaction_time_max - reaction_time_min) / 3:
        return 'Prompt'
    else:
        return 'Delay'

df['react_time'] = df.apply(classify_reaction_time, axis=1)
```

### Retrieving Year, Month and Time from Call Date

```python
# Convert the 'Call Date' column to datetime format
df['Call Date'] = pd.to_datetime(df['Call Date'])

# Create a new column called 'ret_yr_mnth' with the year and month values
df['ret_yr_mnth'] = df['Call Date'].dt.strftime('%Y-%m')

# Print the first 10 rows of the DataFrame with the new column
print(df.head(10))
```

```python
df['cd_Year'] = df['Call Date'].dt.year
df['cd_Month'] = df['Call Date'].dt.month
df.drop('ret_yr_mnth', axis=1, inplace=True)
```

```python
# Convert the datetime column to datetime format
df['Received DtTm'] = pd.to_datetime(df['Received DtTm'], format='%m/%d/%Y %I:%M:%S %p')

# Create a new column called 'am_pm' with the AM/PM component
df['am_pm'] = df['Received DtTm'].dt.strftime('%p')

# Print the first 10 rows of the DataFrame with the new column
print(df.head(10))
```

### Filling Station Area 

```python
city_sa_mapping = df.dropna(subset=['Station Area']).groupby('City')['Station Area'].agg(pd.Series.mode).to_dict()
def fill_sa(row, mapping):
    if pd.isnull(row['Station Area']):
        return mapping.get(row['City'], np.nan)
    else:
        return row['Station Area']

df['Station Area'] = df.apply(lambda row: fill_city(row, city_sa_mapping), axis=1)
```
