Response DtTm	: Date & Time :	Date and time this unit acknowledges the dispatch and records that the unit is en route to the location of the call.

On Scene DtTm : Date & Time :	Date and time the unit records arriving to the location of the incident.

The missing values of Response DtTm and On Scene DtTm will be filled with Entry DtTm as it is Date and time the 911 operator submits the entry of the initical call information into the CAD system which can be considered to fill the nan values 


```python
df['Response DtTm'].fillna(df['Entry DtTm'], inplace=True)
df['On Scene DtTm'].fillna(df['Entry DtTm'], inplace=True)
```

```python
#df['Transport DtTm'].head(25)
```

Transport DtTm 	: Date & Time :	If this unit is an ambulance, date and time the unit begins the transport unit arrives to hospital

Hospital DtTm  : Date & Time :	If this unit is an ambulance, date and time the unit arrives to the hospital..

The missing values of Transport DtTm and On Hospital DtTm will be filled with Dispatch DtTm DtTm as it is Date and time the 911 operator dispatches this unit to the call which can be considered to fill the nan values

```python
df['Transport DtTm'].fillna(df['Dispatch DtTm'], inplace=True)
df['Hospital DtTm'].fillna(df['Dispatch DtTm'], inplace=True)
```

Priority and Original Priority are alphanumeric values. So, NaN values can be replaced with mode

```python
mode_value_priority=df['Priority'].mode()
print(mode_value_priority)
mode_value_og_priority=df['Original Priority'].mode()
print(mode_value_og_priority)
```

```python
df['Priority'].fillna(value=mode_value_priority, inplace=True)
print(df['Priority'].head(5))
df['Original Priority'].fillna(value=mode_value_og_priority, inplace=True)
print(df['Original Priority'].head(5))
```
